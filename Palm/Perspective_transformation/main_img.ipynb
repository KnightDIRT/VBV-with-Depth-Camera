```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread("1978.png")
pic_RGB = cv2.cvtColor(pic,cv2.COLOR_BGR2RGB)
plt.imshow(pic_RGB)
```

```python
#LowerLeft = 0,382
#LowerRight = 640,382
#UpperLeft = 120,283
#UpperRight = 520,283
origin_points = [[200,200],[440,200],[0,240],[640,240]]
target_points = [[0,0],[350,0],[0,400],[350,400]]

M1 = cv2.getPerspectiveTransform(np.float32(origin_points),np.float32(target_points))
print(M1)

res = cv2.warpPerspective(pic_RGB,M1,(350,400))

plt.imshow(res)
```
